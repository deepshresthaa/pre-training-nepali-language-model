```python
import numpy as np
import matplotlib.pyplot as plt

losses = np.load("./checkpoints/epoch_3.npz")["meta/train_loss_history"]

print("TOTAL:", len(losses))

steps = np.arange(len(losses))

plt.figure(figsize=(10, 5))
plt.plot(steps, losses)

plt.xlabel("Training Step")
plt.ylabel("Loss")

plt.title("Epoch (1-3) Training Loss")
plt.grid(True)

plt.show()
```
